```python
import numpy as np
import pandas as pd
import os
import datetime
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analysis_tool = SentimentIntensityAnalyzer()
```

```python
metadata = pd.read_csv("tweetmetadata/Company_Tweet.csv")
tweets = pd.read_csv("actualtweets/Tweet.csv")
```

```python
tweets = pd.merge(metadata, tweets, on='tweet_id', how='inner')
```

```python
tweet_body = tweets["body"]
tweet_body_sample = tweet_body[:10]
tweet_body_sample1=pd.DataFrame(tweet_body_sample)
tweet_body_sample1
```

```python
def negativityCheck(sentence):
    if(analysis_tool.polarity_scores(sentence)["neg"]>0.7):
        return 1
    else:
        return 0
```

```python
tweets.shape
```

```python
tweets["negative_sentiment"] = tweet_body_sample1["negative"]=pd.DataFrame(tweet_body.apply(negativityCheck).tolist())
```

```python
negativeTweets = tweets.loc[tweets["negative_sentiment"]==1]
negativeTweets.shape
```

```python
tweets.ticker_symbol.unique()
```

```python
tesla_df = pd.read_csv('stockdata/1629679250303.csv')
```

```python
tesla_df.head(2)
```

```python
tesla_df.shape
```

```python
negativeTweetstsla = negativeTweets.loc[negativeTweets['ticker_symbol'] =="TSLA"]
```

```python
negativeTweetstsla.head()
```

```python
negativeTweetstsla['Date'] = pd.to_datetime(negativeTweetstsla['post_date'],unit='s').dt.strftime('%d/%m/%Y')
```

```python
negativeTweetstsla.head(10)
```

```python
tsla_tweet_effect= pd.merge(negativeTweetstsla,tesla_df,on="Date", how='inner')
```

```python
tsla_tweet_effect.head()
```

```python
date_count = tsla_tweet_effect["Date"].value_counts()
```

```python
date_count = pd.DataFrame(date_count)
date_count.reset_index(inplace=True)
```

```python
date_count["count"] = date_count["Date"]
date_count.drop("Date",axis=1,inplace=True)
date_count["Date"] = date_count["index"]
```

```python
tsla_tweet_effect1= pd.merge(date_count,tesla_df,on="Date",how="inner")
```

```python
tsla_tweet_effect1.tail(1)
```

```python
tsla_tweet_effect1.head(1)
```

```python
tsla_tweet_effect1.columns
```

```python
tsla_tweet_effect1.set_index("Date",inplace=True)
```

```python
tsla_tweet_effect1.columns = tsla_tweet_effect1.columns.str.replace(' ', '')
```

```python
tsla_tweet_effect1
```

```python
x_axis = tsla_tweet_effect1['Open']
y_axis = tsla_tweet_effect1['index']
```

```python
plt.ylabel('Date')
plt.xlabel('Open')
figure(figsize=(20, 12), dpi=80)
plt.plot(x_axis,y_axis)
```
